```python
import pandas as pd

df = pd.read_csv("pizzas.csv")
df
```

```python
df
df.plot(kind="scatter", x="diametro", y="preco");
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["diametro"]]
y = df[["preco"]]

modelo.fit(x, y)
```

```python
print(modelo.predict([[30]])[0][0]) # Deve retornar 75.0
print(modelo.predict([[100]])[0][0]) # Deve retornar 250
```
